### Группировки таблиц `pd.DataFrame`

4.1 Посчитайте количество рецептов, представленных каждым из участников (`contributor_id`). Какой участник добавил максимальное кол-во рецептов?

```python
contributor_recipe_counts = recipes['contributor_id'].value_counts()
max_contributor_id = contributor_recipe_counts.index[0]
print(f'Contributor {max_contributor_id} added the maximum number of recipes')
```

4.2 Посчитайте средний рейтинг к каждому из рецептов. Для скольких рецептов отсутствуют отзывы? Обратите внимание, что отзыв с нулевым рейтингом или не заполненным текстовым описанием не считается отсутствующим.

```python
reviewed_recipes = pd.merge(recipes, reviews, left_on='id', right_on='recipe_id', how='left')
no_review_count = reviewed_recipes['review'].isnull().sum()
total_recipes = len(recipes)
print(f'Total recipes: {total_recipes}')
print(f'Recipes with no reviews: {no_review_count}')
print(f'Percentage of recipes with no reviews: {no_review_count / total_recipes * 100:.2f}%')
```

5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
review_counts = reviewed_recipes.groupby('id')['review'].count()
review_counts.name = 'review_count'
review_counts = review_counts.reset_index()
recipes_with_reviews = pd.merge(recipes, review_counts, left_on='id', right_on='id', how='left')
recipes_with_reviews['review_count'] = recipes_with_reviews['review_count'].fillna(0)
recipes_with_reviews = recipes_with_reviews[['id', 'name', 'review_count']]
print(recipes_with_reviews.head())
```

5.3. Выясните, рецепты, добавленные в каком году, имеют наименьший средний рейтинг?

```python
reviewed_recipes = pd.merge(recipes, reviews, left_on='id', right_on='recipe_id', how='left')
mean_ratings_by_year = reviewed_recipes.groupby('year')['rating'].mean()
worst_year = mean_ratings_by_year.idxmin()
print(f'The year with the worst average rating is {worst_year}')
```
